```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/dados_ibge.csv', sep=';')
df.head()
```

```python
df.info()
```

```python
corr = df.corr()
sns.set(rc = {'figure.figsize':(25,20)})
ax = sns.heatmap(corr, annot=True)
```

```python
cols = ['municipio', 'densidade_demografica', 'populacao_estimada', 'salario_medio_mensal_trabalhadores_formais', 'populacao_ocupada', 'perc_pop_rend_per_capta_meio_sm', 
        'matriculas_ensino_fundamental', 'matriculas_ensino_medio', 'tx_escolarizacao_16_14', 'IDEB_anos_iniciais_ef', 'IDEB_anos_finais_ef', 'escolas_ensino_fundamental', 'escolas_ensino_medio', 'idhm',
        'pib_per_capta', 'qt_estabelecimentos_saude_sus', 'area_unidade_territorial', 'mortalidade_infantil', 'esgotamento_sanitario_adequado', 'populacao_exposta_risco']

df_filter = df[cols].copy()
```

```python
corr = df_filter.corr()

sns.set(rc = {'figure.figsize':(25,20)})
ax = sns.heatmap(corr, annot=True)
```

```python
# sns.set(rc = {'figure.figsize':(25,20)})
sns.pairplot(df_filter)
```

```python
df_filter['escolas_fund_hab'] = df_filter['escolas_ensino_fundamental']*100000 / df_filter['populacao_estimada']
df_filter['escolas_medio_hab'] = df_filter['escolas_ensino_medio']*100000 / df_filter['populacao_estimada']
df_filter['matriculas_fund_hab'] = df_filter['matriculas_ensino_fundamental']*100000 / df_filter['populacao_estimada']
df_filter['matriculas_fund_hab'] = df_filter['matriculas_ensino_fundamental']*100000 / df_filter['populacao_estimada']
df_filter['sus_hab'] = df_filter['qt_estabelecimentos_saude_sus']*100000 / df_filter['populacao_estimada']
```

```python
cols = ['municipio', 'densidade_demografica', 'populacao_estimada', 'salario_medio_mensal_trabalhadores_formais', 'populacao_ocupada', 'perc_pop_rend_per_capta_meio_sm', 
        'matriculas_ensino_fundamental', 'matriculas_ensino_medio', 'tx_escolarizacao_16_14', 'IDEB_anos_iniciais_ef', 'IDEB_anos_finais_ef', 'escolas_ensino_fundamental', 'escolas_ensino_medio', 'idhm',
        'pib_per_capta', 'qt_estabelecimentos_saude_sus', 'area_unidade_territorial', 'mortalidade_infantil', 'esgotamento_sanitario_adequado', 'populacao_exposta_risco']
```

```python

fig, axes = plt.subplots(nrows=3, ncols=3)

# df1.plot(ax=axes[0,0])
# df2.plot(ax=axes[0,1])

df_filter[['escolas_fund_hab', 'escolas_medio_hab']].boxplot(ax=axes[0,0])
df_filter[['matriculas_fund_hab', 'matriculas_fund_hab']].boxplot(ax=axes[0,1])
df_filter[['populacao_ocupada', 'tx_escolarizacao_16_14']].boxplot(ax=axes[0,2])
df_filter[['pib_per_capta']].boxplot(ax=axes[1,0])
df_filter[['idhm']].boxplot(ax=axes[1,1])
df_filter[['sus_hab']].boxplot(ax=axes[1,2])
df_filter[['populacao_exposta_risco']].boxplot(ax=axes[2,0])
df_filter[['esgotamento_sanitario_adequado']].boxplot(ax=axes[2,1])
df_filter[['mortalidade_infantil']].boxplot(ax=axes[2,2])
```
